Estatísticas de todos os jogadores de futebol nas 5 principais ligas 24/25 - Projeto 3

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("orkunaktas/all-football-players-stats-in-top-5-leagues-2425")

print("Path to dataset files:", path)
```

Visualizar as primeiras linhas

```python
import os

# Listar arquivos no diretório baixado
print(os.listdir(path))
```

Carregar arquivo do Excel

```python
import pandas as pd

# Definir o caminho correto do arquivo
file_path = f"{path}/top5-players24-25.xlsx"

# Carregar a planilha do Excel
df = pd.read_excel(file_path)

# Exibir as primeiras linhas para inspeção
print(df.head())
```

Verificar estrutura do DataSet

```python
# Exibir informações sobre o dataset
print(df.info())

# Verificar valores ausentes
print(df.isnull().sum())

# Estatísticas gerais
print(df.describe())
```

Melhores jogadores Ranking geral

```python
# Filtrar os jogadores com mais contribuições ofensivas
top_jogadores = df.sort_values(by="G+A", ascending=False).head(10)

# Exibir os melhores jogadores
print("\nTop 10 jogadores com mais contribuições ofensivas (Gols + Assistências):")
print(top_jogadores[['Player', 'Squad', 'Comp', 'Gls', 'Ast', 'G+A']])
```

Artilheiros

```python
# Filtrar os jogadores com mais gols marcados
top_goleadores = df.sort_values(by="Gls", ascending=False).head(10)

# Exibir os maiores goleadores
print("\nTop 10 artilheiros:")
print(top_goleadores[['Player', 'Squad', 'Comp', 'Gls']])
```

Gráfico dos artilheiros

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x="Gls", y="Player", data=top_goleadores, hue="Player", palette="coolwarm", legend=False)
plt.title("Top 10 Artilheiros da Temporada")
plt.xlabel("Gols Marcados")
plt.ylabel("Jogador")
plt.show()
```

Melhores assistentes - jogadores com mais passes para gol

```python
# Filtrar os jogadores com mais assistências
top_assistentes = df.sort_values(by="Ast", ascending=False).head(10)

# Exibir os melhores assistentes
print("\nTop 10 jogadores com mais assistências:")
print(top_assistentes[['Player', 'Squad', 'Comp', 'Ast']])
```

Defensores com mais desarmes

```python
# Verificar se há uma coluna correspondente a desarmes (PrgC pode indicar ações defensivas)
top_defensores = df.sort_values(by="PrgC", ascending=False).head(10)

# Exibir os defensores com mais desarmes
print("\nTop 10 defensores com mais desarmes (PrgC):")
print(top_defensores[['Player', 'Squad', 'Comp', 'PrgC']])
```

Comparação entre as ligas

```python
# Média de gols por liga
media_gols_por_liga = df.groupby("Comp")["Gls"].mean().reset_index()

# Exibir estatísticas
print("\nMédia de gols por jogador em cada liga:")
print(media_gols_por_liga)
```

Gráfico para visualizações

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Comp", y="Gls", data=media_gols_por_liga, hue="Comp", palette="viridis", legend=False)
plt.title("Média de Gols por Liga")
plt.xlabel("Liga")
plt.ylabel("Média de Gols")
plt.xticks(rotation=45)
plt.show()
```

Goleiros com mais defesas e segurança defensiva

```python
print(df.columns)  # Exibir todas as colunas para encontrar métricas defensivas
```

```python
# Identificar os jogadores que mais atuam como goleiros (Pos = 'GK')
goleiros = df[df["Pos"].str.contains("GK", case=False, na=False)]

# Ordenar os goleiros por maior número de defesas (ajuste conforme a coluna disponível)
top_goleiros = goleiros.sort_values(by="PrgR", ascending=False).head(10)

# Exibir os melhores goleiros
print("\nTop 10 goleiros com mais ações defensivas:")
print(top_goleiros[['Player', 'Squad', 'Comp', 'PrgR']])
```

Comparação da idade dos jogadores por liga

```python
# Calcular a idade média dos jogadores por liga
idade_media_por_liga = df.groupby("Comp")["Age"].mean().reset_index()

# Exibir resultados
print("\nIdade média dos jogadores por liga:")
print(idade_media_por_liga)
```

Gráfico para visualização

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Comp", y="Age", data=idade_media_por_liga, hue="Comp", palette="coolwarm", legend=False)
plt.title("Idade Média dos Jogadores por Liga")
plt.xlabel("Liga")
plt.ylabel("Idade Média")
plt.xticks(rotation=45)
plt.show()
```

Analisando os jogadores mais eficientes

```python
# Filtrar os jogadores com maior xG (Expected Goals)
top_xg_jogadores = df.sort_values(by="xG_90", ascending=False).head(10)

# Exibir os melhores jogadores por xG
print("\nTop 10 jogadores com maior eficiência ofensiva baseada em xG:")
print(top_xg_jogadores[['Player', 'Squad', 'Comp', 'xG_90']])
```

Gráfico para visualizar os jogadores mais eficientes

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="xG_90", y="Player", data=top_xg_jogadores, hue="Player", palette="viridis", legend=False)
plt.title("Top 10 Jogadores por Gols Esperados (xG)")
plt.xlabel("Gols Esperados por 90 Minutos")
plt.ylabel("Jogador")
plt.show()
```
